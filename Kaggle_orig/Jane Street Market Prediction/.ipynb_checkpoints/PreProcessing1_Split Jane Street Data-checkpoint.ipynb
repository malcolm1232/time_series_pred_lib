### Since data of Jane Street is 6GB, we want to make it 500MB each so it is readable on Kaggle

```python
# Split data into equal lengths
len_df = len(df)
indiv_len = int(len(df)/12)
indiv_len
# df_num = 1
for i in range(0,13):
    print("i:",i)
    start_len = (i)*indiv_len
    end_len = (i+1)*indiv_len
    print("start_len:",start_len)
    print("end_len:",end_len)
    exec(f'df{i}=df[start_len:end_len]')
#     exec(f'tempdf = df{i}')
#     print(tempdf)
    exec(f'df{i}.to_csv("./Jane Street Data/Splitted_Data/JaneStreet_Part{i}.csv")')
```
